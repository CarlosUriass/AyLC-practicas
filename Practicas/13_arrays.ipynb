```python
import numpy as np

# Declarar array
numeros = np.zeros(10)

# Ingresar valores al array
for i in range(1, 11):
    num = float(input(f"Ingrese el n√∫mero {i}: "))
    numeros[i-1] = num

# Calcular la suma de los elementos del array
suma = np.sum(numeros)

# Mostrar el resultado
print(f"La suma de los datos es: {suma}")
```
